```python
airbnb.drop(columns=['room_id','survey_id','host_id','country','city','borough','reviews', 'name','bathrooms','minstay', 'last_modified', 'latitude', 'longitude', 'location'], inplace=True)
```

```python
# Dropping Null Values

airbnb.dropna(inplace=True)
```

## Univariant Analysis 

#### Room type column

```python
airbnb.room_type.value_counts()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(data=airbnb, x = 'room_type')
plt.show()
```

```python
plt.figure(figsize=(7,5))
airbnb.room_type.value_counts().plot(kind='pie', autopct='%1.2f%%')
plt.show()
```

#### Neighbourhood column 

```python
airbnb.neighborhood.value_counts()
```

```python
plt.figure(figsize=(15,8))
sns.countplot(data=airbnb, x = 'neighborhood')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(15,15))
airbnb.neighborhood.value_counts().plot(kind='pie', autopct='%1.2f%%')
plt.show()
```
